```python
def trova_distribuzione(country_code):
    random.seed(1234)
    results = list()
    # country_code unique, valori non nulli
    values = pd.unique(country_code.dropna())
    # append questo valore alla lista risultato
    results.append(values)
    # probabilità di occorrenza
    probs = country_code.dropna().value_counts() / len(country_code.dropna())
    # append questa probabilità alla lista risultato
    results.append(probs)
    # indice delle righe dove country code è null
    index_rows_nan = country_code.index[country_code.isnull()]
    # append questo indice alla lista risultato
    results.append(index_rows_nan)
    # n lunghezza del vettore Nan (numero di righe che hanno country code Nan)
    n = len(index_rows_nan)
    #append these value to results list
    results.append(n)
    return (results)
```

```python
# unique country code non nulli
# proababilità di occorrenza per country code
# indice delle righe dove country code è nullo
# numero di righe che hanno country code Nan

risultato_funzione = trova_distribuzione(lenders['country_code'])
risultato_funzione
```

```python
distribuzione_calcolata_country_code = pd.Series(np.random.choice(risultato_funzione[0], p = risultato_funzione[1],
                                                size = risultato_funzione[3]), index = risultato_funzione[2])
distribuzione_calcolata_country_code

# ritorna una lista dei contry_code che ERANO NULLI e adesso sono stati rimpiazzati
```

```python
# inserisco nel dataset
lenders['country_code'].fillna(distribuzione_calcolata_country_code, inplace = True)
lenders

# ora lenders ha tutti contry_code
```

## 12. Per ogni anno calcolare l'importo totale dei *loans* (distribuiti proporzionalmente al numero di giorni all'anno) 

#### Per ogni anno, calcolare l'importo totale dei prestiti. Ogni prestito che ha pianificato il tempo di scadenza e il tempo di erogazione in anni diversi deve avere il suo importo distribuito proporzionalmente al numero di giorni di ogni anno. Ad esempio, un prestito con tempo di erogazione 1° dicembre 2016, tempo di scadenza previsto 30 gennaio 2018 e importo 5000USD ha un importo di 5000USD * 31 / (31+365+30) = 363,85 per il 2016, 5000USD * 365 / (31+365+30) = 4284,04 per il 2017, e 5000USD * 30 / (31+365+30) = 352,11 per il 2018.


```python
# funzione per ottenere il numero di giorni tra due date

def diff_giorni(data1, data2):
    
    # controllo di evitare numeri negativi --> ordinare la coppia di date
    (inizio, fine) = (data1, data2) if data1<data2 else (data2,data1)
    
    # lista dove: prima data, ogni primo gennaio... data di fine 
    data_list = list()
    
    # il primo elemento della lista è la data di inizio 
    data_list.append(inizio)
    
    # per ogni anno tra la data di inizio e la data di fine aggiungi ogni 1° gennaio
    # degli anni tra la data di inizio e la data di fine
    for a in range (inizio.year+1, fine.year+1):
        data_list.append(pd.Timestamp(year=a, month=1, day=1, tz='utc')) # 1 gennaio dell'anno "a"
        
    # aggiungi data di fine
    data_list.append(fine)
    
    # crea un dizionario
    res = dict()
    
    # anno= chiave
    # il valore è la differnza di giorni tra anno corrente e anno successivo
    for i in range(len(data_list)-1):
        res[data_list[i].year]=(data_list[i+1] - data_list[i]).days
        
    # creare la serie in cui l'indice è la chiave
    return pd.Series(res, name='Valore_data')
    
```

```python
# ora servono gli attributi 
loans[['disburse_time', 'planned_expiration_time']]
```

```python
# to datetime
loans['planned_expiration_time'] = pd.to_datetime(loans['planned_expiration_time']) 
loans['disburse_time'] = pd.to_datetime(loans['disburse_time'])

# aggiungo duration
# numero di giorni che intercorrono tra 
# il disburse_time e il planned_expiration_time 

loans[['disburse_time', 'planned_expiration_time', 'duration']]
```

```python
sum(pd.isnull(loans['duration']))
```

```python
# elimino righe con duration NAN
loans_drop = loans.dropna(subset=['duration'])
```

```python
# loans_drop dovrebbe essere 1419607 - 374645 = 1044962
loans_drop[['disburse_time', 'planned_expiration_time', 'duration']]
```

```python
# durata del loan in giorni per ogni anno

giorni_loan_anno = loans_drop.apply(
    lambda x: diff_giorni(x['planned_expiration_time'], x['disburse_time']),
                          axis=1)

giorni_loan_anno
```

```python
# replace Nan con 0
giorni_loan_anno.fillna(0, inplace = True)
```

```python
# concat (dal dataset loans_drop) --> loan_id, loan_amount, duration

giorni_loan_anno_finish = pd.concat([giorni_loan_anno, 
                              loans_drop['loan_id'],
                              loans_drop['loan_amount'],
                              loans_drop['duration'] ],
                             axis = 1, sort = True)

giorni_loan_anno_finish
```

```python
importo_proporzionale = giorni_loan_anno_finish[[2011,2012,2013,2014,2015,2016,2017,2018
                        ]].div(giorni_loan_anno_finish.duration,
                            axis=0).mul(giorni_loan_anno_finish.loan_amount,
                                        axis = 0)

importo_proporzionale
```

```python
# trasponi
importo_proporzionale = importo_proporzionale.transpose()

# somma righe 
df = importo_proporzionale["importo_totale_dei_loans"] = importo_proporzionale.sum(axis=1)
df = df.reset_index(name='importo_totale')

# per togliere notaz scientif
pd.set_option('display.float_format', lambda x: '%.5f' % x) 

df
```
